### Merging

```python
d1=np.array([[1,2,3],[4,5,6]])
d1=pd.DataFrame(d1)
d1
```

```python
d2=np.array([[7,8,9],[10,11,12],[13,14,15]]);
d2=pd.DataFrame(d2)
d2.info()
```

```python
pd.concat([d1,d2],axis=0) # axix=0 --> col concat
```

```python
pd.concat([d1,d2],axis=1,ignore_index=True) # ingone_inder --> removes the previous index with new one
```

```python
d1=pd.DataFrame({
    "City":['Lucknow','Kanpur','Agra','Delhi'],
    "Temperature":[35,45,30,40]
})
d1
```

```python
d2=pd.DataFrame({
    "City":['Delhi',"Lucknow",'Kanpur'],
    "Temperature":[68,65,75]
})
d2
```

```python
pd.merge(d1,d2,on='City')
```

```python
pd.merge(d1,d2,on="City",how='left') 
```
