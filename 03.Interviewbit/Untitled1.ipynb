```python
import time
import pyautogui
import webbrowser
import pandas as pd
```

```python
facebook = pd.read_excel('/home/anuj/Downloads/facebook (1).xlsx')
facebook
```

```python
groups_list = list(facebook['NFT Group link'])
groups_list
```

```python
def split(list_a, chunk_size):

  for i in range(0, len(list_a), chunk_size):
    yield list_a[i:i + chunk_size]

chunk_size = int(len(groups_list)/5)
group_list_01=list(split(groups_list, chunk_size))
group_list_01
```

```python
b=webbrowser.get('firefox')
for url in group_list_01[0]:
    b.open(url)
    print("Waiting for 5 seconds\n")
    time.sleep(5)
    pyautogui.typewrite('')
    time.sleep(2)
    print("Writing post\n")
    pyautogui.typewrite("https://opensea.io/collection/hanumaan")
    time.sleep(4)
    pyautogui.keyDown('ctrl')
    pyautogui.keyDown('enter')
    pyautogui.keyUp('enter')
    pyautogui.keyUp('ctrl')

    time.sleep(3)

    pyautogui.write(['f6'])
    time.sleep(1)
```
