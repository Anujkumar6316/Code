```python
# adding columns to dataframe
seats=pd.Series([2,3,4,5])
car_sales_DF['Seats']=seats
car_sales_DF
```

```python
fuel_ussage=[2,4,5,9,4,6,7,8,2,1]
car_sales_DF['Fuel per 100km']=fuel_ussage
car_sales_DF
```

```python
car_sales_DF['Total fuel used(L)']=car_sales_DF['Odometer (KM)']/100*car_sales_DF['Fuel per 100km']
car_sales_DF
```

```python
# dropping collumn in df
# here we can also use inplace 

car_sales_DF=car_sales_DF.drop('Seats',axis=1) #axis=1 for column
car_sales_DF
```

```python
car_sales_shuffled_df=car_sales_DF.sample(frac=0.8)
car_sales_shuffled_df
```

```python
car_sales_shuffled_df.reset_index(drop=True, inplace=True)
car_sales_shuffled_df
```
